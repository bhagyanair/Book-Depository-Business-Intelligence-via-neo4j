# VII. Null value handling 

```python
# checking null values in the main dataset
pd.isnull(data).sum()
```

Imputation with median for numerical attributes. 

Customised 'missing information' message for categorical attributes.

```python
data['bestsellers-rank']= data['bestsellers-rank'].fillna(0)
data['description']= data['description'].fillna('No description available')
data['dimension-x']= data['dimension-x'].fillna(0)
data['dimension-y']= data['dimension-y'].fillna(0)
data['dimension-z']= data['dimension-z'].fillna(0)
data['edition']= data['edition'].fillna('None')
data['edition-statement']= data['edition-statement'].fillna('Not available')
data['for-ages']= data['for-ages'].fillna('Not available')
data['format']= data['format'].fillna(0)
data['illustrations-note']= data['illustrations-note'].fillna('Not available')
data['image-checksum']= data['image-checksum'].fillna(0)
data['image-path']= data['image-path'].fillna(0)
data['image-url']= data['image-url'].fillna(0)
data['imprint']= data['imprint'].fillna('Not available')
data['lang']= data['lang'].fillna('Not available')
data['publication-date']= data['publication-date'].fillna(0)
data['rating-avg']= data['rating-avg'].fillna(data['rating-avg'].mean(),)
data['rating-count']= data['rating-count'].fillna(data['rating-count'].mean(),)# fill with mean value
# data['rating-avg']= data['rating-avg'].fillna(((data['rating-avg'].mean())))
# data['rating-count']= data['rating-count'].fillna(((data['rating-count'].mean())))
data['weight']= data['weight'].fillna(0)
```
